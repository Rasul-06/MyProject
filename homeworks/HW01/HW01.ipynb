```python
# Импортируем библиотеку NumPy с общепринятым сокращением np
import numpy as np
```

```python
# Создаём массив из списка Python
x = np.array([5, 10, 15, 20])
x
```

```python
# Массив, заполненный нулями
zeros_array = np.zeros((3, 2))  # 3 строки, 2 столбца
zeros_array
```

```python
# Массив, заполненный единицами
ones_array = np.ones(6)
ones_array
```

```python
# Массив с равномерной последовательностью чисел
sequence = np.arange(1, 11, 3)  # от 1 до 10 с шагом 3
sequence
```

```python
# Проверим форму и тип данных массива x
print("Форма массива x:", x.shape)   # (4,) — одномерный массив с 4 элементами
print("Тип данных массива x:", x.dtype)  # int64
```

```python
# Получение одного элемента массива
print("Третий элемент массива x:", x[2])
```

```python
# Получение среза массива
print("Срез массива sequence (элементы 2-го и 3-го):", sequence[1:3])
```

```python
# Создаём второй массив той же длины
y = np.array([2, 4, 6, 8])

# Поэлементное сложение
print("Сложение x + y:", x + y)
```

```python
# Умножение на число
print("Умножение массива x на 5:", x * 5)
```

```python
# Вычисление средней величины массива y
print("Среднее значение массива y:", y.mean())
```
